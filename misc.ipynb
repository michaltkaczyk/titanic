```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns

sns.set_theme(style = "dark")
```

```python
titanic_df = pd.read_csv("titanic.csv")
```

`pandas.DataFrame.hist` vs `seaborn.displot`

```python
titanic_df["Age"].hist(bins = 40).set(title = "Number of passengers by age", xlabel = "Age", ylabel = "Number of passengers")
```

```python
sns.displot(titanic_df, x = "Age", bins = 40).set(
    title = "Number of passengers by age",
    xlabel = "Age",
    ylabel = "Number of passengers")
```
